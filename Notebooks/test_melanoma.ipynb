```python
x_pixels = []
y_pixels = []

for spot in df_man.index:
    y_pixels.append(int(spot.split('x')[1]))
    x_pixels.append(int(spot.split('x')[0]))
```

```python
plt.scatter(y_pixels, x_pixels, c=list(map(lambda x: color_map[x], df_man['label'])))
```
